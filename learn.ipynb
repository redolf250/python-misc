```python
import os
import json
import subprocess
from pathlib import Path
```

```python
base_directory = r"D:\Projects\BizBoard\BizBoard.Mvc"
extensions = [".cs", ".cshtml"]
```

```python
def find_files_by_extensions(base_dir, extensions):
    base_path = Path(base_dir)
    return [str(file) for file in base_path.rglob("*") if file.suffix in extensions]
    
files = find_files_by_extensions(base_directory, extensions)
```

```python
def check_git_installation():
    try:
        result = subprocess.run(['git', '--version'], capture_output=True, text=True)
        print(result.stdout)
        return result.stdout
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")
        return f"An error occurred: {e}"
```

```python
datastore={}
def data_store(file_name: str, changes: str):
    datastore[file_name] = changes
    return datastore
```

```python
def get_file_line_changes_dir_walk_V2(file_name):
    print(f"Processing: {file_name}:")
    result = subprocess.run(['git', 'diff', file_name], capture_output=True, text=True)
    if result.returncode == 0:
        changes = result.stdout.strip()
        if changes:
            data_store(file_name, changes)
    else:
        print(f"Error: {result.stderr}")
```

```python
def get_file_changes(files: list):
    for item in range(len(files)):
        os.chdir(base_directory)
        get_file_line_changes_dir_walk_V2(files[item])
get_file_changes(files)
```

```python
def git_commit_changes(file: str, message: str):
    try:
        subprocess.run(['git', 'add', file], check=True)
        subprocess.run(['git', 'commit', '-m', message], check=True)
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")
```

```python
def git_commit(message: str):
    try:
        process = subprocess.Popen(['git', 'commit', '-m', 'Initial commit'], stdin=subprocess.PIPE, text=True)
        process.communicate()
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")  
```
